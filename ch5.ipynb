```python
def Change(money,change):
  mincoin=(money+1)*[0]
  for m in range(1,money+1):
    mincoin[m]=m
    for coin in change:
      if m >= coin:
        if mincoin[m-coin]<mincoin[m]:
          mincoin[m]=mincoin[m-coin]+1
  return mincoin[money]
ans = Change(40, [1,5,10,20,25,50])
print(ans)
```

```python
def mannhatantourist(n,m,down,right):
  s=[]
  for i in range(n+1):
    s.append((m+1)*[0])

  for i in range(1,n+1):
    s[i][0]=s[i-1][0]+down[i-1][0]

  for j in range(1,m+1):
    s[0][j]=s[0][j-1]+right[0][j-1]

  for i in range(1,n+1):
    for j in range(1,m+1):
      s[i][j]=max(s[i-1][j]+down[i-1][j],s[i][j-1]+right[i][j-1])
  return s[n][m]
ans = mannhatantourist(4, 4, [[1, 0, 2, 4, 3],
                              [4, 6, 5, 2, 1],
                              [4, 4, 5, 2, 1],
                              [5, 6, 8, 5, 3]],
                             [[3, 2, 4, 0],
                              [3, 2, 4, 2],
                              [0, 7, 3, 3],
                              [3, 3, 0, 2],
                              [1, 3, 2, 2]])

print(ans)
```

```python
def LCS(s1,s2):
  l1=len(s1)
  l2=len(s2)
  m=[]
  for i in range(l1+1):
    m+=[[0]*(l2+1)]
  for i in range(1,l1+1):
    for j in range(1,l2+1):
      if s1[i-1]==s2[j-1]:
        m[i][j]=m[i-1][j-1]+1
      else:
        m[i][j]=max(m[i-1][j],m[i][j-1])
  lcs_length = m[l1][l2]
  lcs = []

  i = l1
  j = l2

  while i > 0 and j > 0:
    if s1[i - 1] == s2[j - 1]:
      lcs.append(s1[i - 1])
      i -= 1
      j -= 1
    elif m[i - 1][j] > m[i][j - 1]:
      i -= 1
    else:
      j -= 1

  return "".join(lcs[::-1])
LCS('AACCTTGG', 'ACACTGTGA')
```

```python
with open("rosalind_ba5c.txt", "r", encoding='utf-8') as f:
  inlines = [x.strip("\n") for x in f.readlines()]
  s1 = inlines[0]
  s2 = inlines[1]

LCS(s1, s2)
```

```python
def editDistance(s1,s2):
  m={}
  for j in range(len(s1)+1):
    m[j,0]=j
  for i in range(len(s2)+1):
    m[0,i]=i
  for j in range(len(s1)):
    for i in range(len(s2)):
      if s2[i]==s1[j]:
        m[j+1,i+1]=m[j,i]
      else:
        m[j+1,i+1]=min([m[j+1,i],m[j,i],m[j,i+1]])+1
  return m[len(s1),len(s2)]
editDistance('PLEASANTLY','mEANLY')
```

```python
with open("rosalind_ba5g.txt", "r", encoding='utf-8') as f:
  inlines = [x.strip("\n") for x in f.readlines()]
  s1, s2 = inlines

editDistance(s1, s2)
```
