```python
def format_num(x):
  if x>=0:
    return f"+{x}"
  else:
    return f"{x}"

def rosalind_print_perm(permutations):
  strings=[]
  for perm in permutations:
    strings.append("("+" ".join(format_num(x) for x in perm)+")")

  return "\n".join(strings)

def greedy_sorting(str_permutations):
  helper=[int(x) for x in str_permutations[1:-1].split()]
  S=[]
  for i in range(0,len(helper)):
    if helper[i]==i+1:
      continue
    idx=i
    while True:
      if helper[idx]==i+1 or helper[idx]==-1*(i+1):
        break

      idx+=1
    mid=[-1*x for x in helper[i:(idx+1)]][::-1]
    helper=helper[0:i]+mid+helper[(idx+1):]
    S.append(helper.copy())
    if helper[i]<0:
      helper[i]=abs(helper[i])
      S.append(helper.copy())
  return S
rosalind_print_perm(greedy_sorting('(-3 +4 +1 +5 -2)'))
```

```python
with open("rosalind_ba6a.txt", "r", encoding='utf-8') as f:
  inlines = [x.strip("\n") for x in f.readlines()]
res = greedy_sorting(str(inlines[0]))
res_output= rosalind_print_perm(res)
res_output

```
