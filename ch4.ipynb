```python
with open('/content/drive/MyDrive/bioalg/gen_code.txt', 'r') as f:
  gen_code = [line.strip().split() for line in f.readlines()]

def get_genetic_code(reverse=False):
  if reverse:
    G={}
    for pair in gen_code:
      if pair[1] not in G:
        G[pair[1]]=[pair[0]]
      else:
        G[pair[1]].append(pair[0])
    return G
  else:
    G={}
    for pair in gen_code:
      G[pair[0]]=pair[1]
    return G
```

```python
with open('/content/drive/MyDrive/bioalg/mass.txt', 'r') as f:
  lines = [line.strip().split(':') for line in f.readlines()]
def get_mass():
  G = {}
  for pair in lines:
    G[pair[0]] = int(pair[1])
  return G
```

```python
def revcomp(string):
  comp=""
  for elem in string:
    if elem=="A":
      comp+="T"
    if elem=="T":
      comp+="A"
    if elem=="G":
      comp+="C"
    if elem=="C":
      comp+="G"
  return comp[::-1]
def genomSubstring(rna,pattern):
  amino=get_genetic_code(True)
  G=get_genetic_code()
  res=[]
  output=[]
  substring=""
  rna=rna.replace("U","T")
  for j in range(0,len(rna),3):
    for i in range(len(res),len(pattern)):
      pat=amino[pattern[i]]
      if rna[j:j+3] in pat:
        res.append(rna[j:j+3])
        substring+=rna[j:j+3]
      if len(res)==len(pattern):
        output.append(substring)

        res.clear()
        reverscomp=revcomp(substring)
        substring=""

        if reverscomp in rna[j-1:]:
          output.append(reverscomp)

  return "\n".join(output)



genomSubstring("ATGGCCATGGCCCCCAGAACTGAGATCAATAGTACCCGTATTAACGGGTGA","MA")
```

```python
def spectrum(peptide):
  n=len(peptide)
  mass=get_mass()
  extended_peptide=peptide+peptide[:-2]
  spectrum=[0,sum([mass[x] for x in peptide])]
  for l in range(n):
    for k in range(1,n):
      subpeptide=extended_peptide[l:l+k]
      spectrum.append(sum([mass[x] for x in subpeptide]))
  spectrum.sort()
  return ' '.join(str(x) for x in spectrum)
peptide ='LEQN'
spectrum(peptide)
```

#BA4D

```python
def getallcombinationsWorker(mass,aa_masses,counter):
  if mass in counter:
    return counter[mass]
  if mass<0:
    return 0
  if mass==0:
    return 1
  counter[mass]=0
  for aa_mass in aa_masses:
    counter[mass]+=getallcombinationsWorker(mass-aa_mass,aa_masses,counter)
  return counter[mass]

def getallCombinations(mass):
  aa_masses=set(get_mass().values())
  counter={}
  res=getallcombinationsWorker(mass,aa_masses,counter)
  return res

getallCombinations(1024)
```

#BA4F

```python
def findSpectrum(peptide):
  n=len(peptide)
  mass=get_mass()
  extended_peptide=peptide+peptide[:-2]
  spectrum=[]
  spectrum.append(0)
  spectrum.append(sum([mass[x] for x in peptide]))
  for l in range(n):
    for k in range(1,n):
      subpeptide=extended_peptide[l:l+k]
      spectrum.append(sum(mass[x]for x in subpeptide))
  return sorted(spectrum)

def Score(peptide,spectrum_):
  pep_spec=findSpectrum(peptide)
  result=0
  unique_masses=set(pep_spec+spectrum_)
  for mass in unique_masses:
    result+=min(pep_spec.count(mass),spectrum_.count(mass))
  return result

Score('NQEL',[0, 99, 113, 114, 128, 227, 257, 299, 355, 356, 370, 371, 484])
```

```python
with open("rosalind_ba4f.txt", "r") as f:
  peptide = f.readline().strip()
  spect = f.readline().split()
spectrum_ = [int(x) for x in spect]
Score(peptide, spectrum_)
```

#BA4H

```python
def convolution(spectrum):
  spectrum.sort()
  conv=[]
  for i in range(len(spectrum)-1):
    for j in range(i,len(spectrum)):
      if spectrum[j]-spectrum[i]!=0:
        conv.append(spectrum[j]-spectrum[i])
  freq_dict={}
  for mass in set(conv):
    freq_dict[mass]=conv.count(mass)
  sorted_mass_list=[k for k, _ in sorted(freq_dict.items(),key=lambda item:item[1],reverse=True)]
  conv=[]
  for mass in sorted_mass_list:
    conv+=[mass]*freq_dict[mass]
  return ' '.join(str(x) for x in conv)

convolution([0, 137, 186, 323])
```

```python
with open("rosalind_ba4h.txt", "r") as f:
  spect = f.readline().split()
spectrum_ = [int(x) for x in spect]
res = convolution(spectrum_)
res
```

#BA4J

```python
def linearSpectrum(peptide,external=False):
  masses=get_mass()
  spectrum=[0]
  n=len(peptide)
  for l in range(n):
    for k in range(1,n-l+1):
      subpeptide=peptide[l:l+k]
      spectrum.append(sum(masses[x] for x in subpeptide))
  spectrum.sort()
  if external:
    return spectrum
  return ' '.join(str(x) for x in spectrum)

peptide = 'NQEL'
linearSpectrum(peptide)
```

```python
with open("rosalind_ba4j.txt", "r") as f:
  peptide = f.readline().strip()
res = linearSpectrum(peptide)
res
```

#BA4K

```python
def linearscore(peptide,spectrum):
  pep_spec=linearSpectrum(peptide,external = True)
  result=0
  uniqe_mass=set(pep_spec+spectrum)
  for mass in uniqe_mass:
    result+=min(pep_spec.count(mass),spectrum.count(mass))
  return result

linearscore('NQEL', [0, 99, 113, 114, 128, 227, 257,299, 355, 356, 370, 371, 484])
```

```python
with open("rosalind_ba4k (1).txt", "r") as f:
  peptide = f.readline().strip()
  spect = f.readline().split()
spectrum_ = [int(x) for x in spect]
res = linearscore(peptide, spectrum_)
res
```

#BA4L

```python
def Trim(leaderboard,spectrum,N):
  if len(leaderboard)<=N:
    return leaderboard
  scores={}
  for i, peptide in enumerate(leaderboard):
    scores[i]=linearscore(peptide,spectrum)
  sorted_scores=sorted(scores.values(),reverse=True)
  threshold=sorted_scores[N-1]
  return " ".join([leaderboard[i] for i, score in scores.items() if score>= threshold])
ans = Trim(['LAST', 'ALST', 'TLLT', 'TQAS'],
 [0, 71, 87, 101, 113, 158, 184, 188, 259, 271, 372], 2)
print(ans)
```

```python
with open("rosalind_ba4l.txt", "r") as f:
  leaderboard = f.readline().split()
  spect = f.readline().split()
  N = int(f.readline().strip())
spectrum_ = [int(x) for x in spect]
res = Trim(leaderboard, spectrum_, N)
res
```
