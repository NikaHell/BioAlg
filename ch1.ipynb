```python
def hammingDistance(p, q): # VJEZBA 2 - BA1G
  if len(p) != len(q): # ako su duljine razlicite
    return -1 # ne mozemo nastaviti rjesavati problem, vrati -1
  d = 0 # brojac
  for i in range(len(p)): # iteriranje po stringu
    if p[i] != q[i]: # ako su p i q razliciti na istoj poziciji
      d += 1 # povecaj
  return d
def Neighborhood(pattern,d):
  if d==0:
    return {pattern}
  if len(pattern)==1:
    return {"A","T","G","C"}
  neighborhood=[]
  suffixNeighbors=Neighborhood(pattern[1:],d)
  for sufix in suffixNeighbors:

    if hammingDistance(pattern[1:],sufix)<d:
      for nucleotide in ["A","C","G","T"]:
        neighborhood.append(nucleotide+sufix)
    else:
      neighborhood.append(pattern[0]+sufix)

  return neighborhood
Neighborhood("ACG",1)
```
